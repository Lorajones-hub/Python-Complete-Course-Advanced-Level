# GUIs frameworks in Python

As a Python developer, sooner or later you'll want to write applications with graphical user interface (GUI). Luckily, there are lots of Python GUI options available for you.

Here is a link to over 30 cross-platform GUI frameworks in Python: [GUI Programming in Python](https://wiki.python.org/moin/GuiProgramming). 

The most popular GUIs included in all three platforms (Windows, Mac, Linux) are and are Python 3- compatible are: tkinter, appJar, Qt, Gtk, and wxWidgets). 

Choosing a particular toolkit is completely up to you. But in this section we will be working with tkinter GUIs.  

Here is a brief overview for each of the most commonly-used toolkits along with their features so you can compare yourself and decide the one that is right for you!.

### 1. tkinter: 

 - It is Python's standard GUI package. It is always available as it comes with Python when you install it.
 
 
 - The easiest one to get started with.
 
 
 - The most commonly used.
 
 
 - It comes with IDLE which is very useful for experimenting, editing, and debugging with Python.
 
 
 - Provides various controls, such as buttons, labels and text boxes used in a GUI application. These controls are usually called **widgets**.
 
 
### 2. PyQt

- Is a multi-licensed cross-platform GUI toolkit written in C++.


- Has large library of tools and APTs.


- Can be used to develop GUIs for mobile apps.


- Provide many widgets including labels, standard dialogs, toolbars, buttons, text boxes, calendar, etc.


### 3. wxPython

- An open-source C++ Python GUI interface bundled with Python.


- Based on wxWidgets GUI toolkit.


- It includes 2D path drawing API, dockable windows, text editing and word processing widgets, and support many file formats.


- Has a wide range of various widgets, including buttons, check boxes, sliders, and list boxes.

The top GUI interfaces along with their pros and cons can be found here: [Top 7 Python GUI frameworks](https://www.slant.co/topics/6620/~python-gui-frameworks-toolkitsn).

Before you start working with any of these frameworks, you need to install them.
